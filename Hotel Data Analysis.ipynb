**1: Import Libraries and Load Data**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/kaggle/input/heart-analysis/heart.csv")
df = df.drop_duplicates()
print(df.head())
```

**2: Basic Info & Data Quality**

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(f"Duplicates: {df.duplicated().sum()}")
```

**3: Target Variable Distribution**

```python
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Target Distribution: Heart Disease Presence')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

**4: Age Distribution vs Target**

```python
sns.histplot(data=df, x='age', hue='target', bins=30, kde=True, palette='coolwarm')
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**5: Gender vs Heart Disease**

```python
sns.countplot(x='sex', hue='target', data=df, palette='pastel')
plt.title('Heart Disease by Gender (0 = Female, 1 = Male)')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```

**6: Chest Pain Type vs Target**

```python
sns.countplot(x='cp', hue='target', data=df, palette='viridis')
plt.title('Chest Pain Type vs Heart Disease')
plt.xlabel('Chest Pain Type (0-3)')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```

**7: Correlation Heatmap**

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
